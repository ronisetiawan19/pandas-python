# Convert string column into the date type 

```python
import pandas as pd
df = pd.read_csv("weather_data5.csv",parse_dates=['day'])
type(df.day[0])
df
```

# Use date as an index of dataframe usine set_index() method 

```python
df.set_index('day',inplace=True)
df
```

# Use fillna() method in dataframe

Fill all NaN with one specific value

```python
new_df = df.fillna(0)
new_df
```

```python
new_df = df.fillna({
        'temperature': 0,
        'windspeed': 0,
        'event': 'No Event'
    })
new_df
```

# Use fillna(method="ffill") method in dataframe

Use method to determine how to fill na values

```python
new_df = df.fillna(method="ffill")
new_df
```

# Use fillna(method="bfill") method in dataframe

```python
new_df = df.fillna(method="bfill")
new_df
```

# "axis" parameter in fillna() method in dataframe

```python
new_df = df.fillna(method="bfill", axis="columns") # axis is either "index" or "columns"
new_df
```

# "limit" parameter in fillna() method in dataframe

```python
new_df = df.fillna(method="ffill",limit=2)
new_df
```

# interpolate() to do interpolation in dataframe

```python
new_df = df.interpolate()
new_df
```

# interpolate() method "time"

```python
new_df = df.interpolate(method="time") 
new_df
```

**Notice that in above temperature on 2017-01-04 is 29 instead of 30 (in plain linear interpolate)**

**There are many other methods for interpolation such as quadratic, piecewise_polynomial, cubic etc. 
Just google "dataframe interpolate" to see complete documentation**

# dropna() method Drop all the rows which has "na" in dataframe

```python
new_df = df.dropna()
new_df
```

```python
new_df = df.dropna(how='all')
new_df
```

```python
new_df = df.dropna(thresh=2)
new_df
```
